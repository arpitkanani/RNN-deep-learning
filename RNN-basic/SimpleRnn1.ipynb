```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.datasets import imdb
from keras.models import Model,Sequential
from keras.layers import Dense,Flatten,SimpleRNN,Embedding,Dropout
```

```python
data=["hello there?",
      "how are you?",
      "I am fine.",
      "what about you?",
      "nice to hear that.",
      "have a great day!",
      "see you later."]
```

#### if dataset is given like this then first convert it into vector for model 

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(data)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences=tokenizer.texts_to_sequences(data)
sequences
```

```python

padded_sequences=pad_sequences(sequences,padding='post')
padded_sequences
```

## now on imdb dataset which is already on vector form

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train.shape,X_train
```

```python
X_train_padd=pad_sequences(X_train,padding='post')
X_test_padd=pad_sequences(X_test,padding='post')
```

```python
X_train_padd[0]
```

```python
model=Sequential()
model.add(SimpleRNN(32,input_shape=(2494,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
weights=32+32 +32*32
weights
```

```python
model.fit(X_train_padd,y_train,epochs=50,steps_per_epoch=100,validation_steps=100,batch_size=64,validation_data=(X_test_padd,y_test))
```
